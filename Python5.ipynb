```python
import pandas as pd
```

```python
link = 'https://www.czso.cz/documents/10180/32853391/300002164403.xlsx/1fed9fb2-4dbf-4b0f-bfbe-4c2d3202a18f?version=1.1'
data = pd.read_excel(link, header = None, index_col = 0, skiprows = 6)
```

```python
data = data.drop(17, 1)
```

```python
data = data.fillna(method='ffill', axis=1)
```

```python
data = data.drop([1,3,5,7,9,11,13,15],  axis=1)
```

```python
col1 = data.iloc[0]
col2 = data.iloc[2].str.split(pat='/', n = 0).str.get(0)
col3 = data.iloc[3]
columns = [col1, col2, col3]
columns = list(zip(*columns))
data.columns = columns
```

```python
data.columns = columns
data = data.drop(data.index[[0,1,2,3,4,7,36,41,42,43]])
```

```python
data.columns = pd.MultiIndex.from_tuples(data.columns, names=['Země', '', ''])
```
